```python
import rosbag
import numpy as np
import matplotlib.pyplot as plt
```

```python
bag = rosbag.Bag('kinematics.bag')
messages = []
for topic, msg, t in bag.read_messages(topics=['world_kinematics']):
    messages.append(msg)
bag.close()
```

```python
n = len(messages)
n
```

```python
X = np.zeros((n, 3))
for i, message in enumerate(messages):
    X[i, :] = message.data
X
```

```python
X_sparse = X[::5, :]
plt.quiver(X_sparse[:, 0], X_sparse[:, 1], np.cos(X_sparse[:, 2]), np.sin(X_sparse[:, 2]), scale=50)
plt.show()
```
