```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from pathlib import Path
from preprocess import warp_image, normalize_image, preprocess_image
```

```python
VISUALIZATION_PATH = Path("data/visualization")
VISUALIZATION_PATH.mkdir(parents=True, exist_ok=True)

image = cv.imread("data/test.png")
image = warp_image(image, VISUALIZATION_PATH)
cv.imwrite(str(VISUALIZATION_PATH / "mask2.png"), image)
image = preprocess_image(image)
cv.imwrite(str(VISUALIZATION_PATH / "preprocess.png"), image)
image = normalize_image(image)

plt.imshow(image, cmap="gray")
```
